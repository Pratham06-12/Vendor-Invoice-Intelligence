# 📑 Vendor Invoice Intelligence: Exploratory Data Analysis (EDA)
This notebook explores the vendor invoice dataset, performs statistical analysis, and visualizes trends to inform feature engineering.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Setting plotting styles
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

## 1. Load Data

```python
df_invoices = pd.read_csv('../data/vendor_invoice.csv')
df_invoices.head()
```

## 2. Univariate Analysis
Analyze the distribution of total invoice amount and freight costs.

```python
sns.histplot(df_invoices['freight_cost'], kde=True, color='blue')
plt.title('Distribution of Freight Costs')
plt.show()
```

## 3. Bivariate Analysis
Relationship between total amount and freight cost.

```python
sns.scatterplot(data=df_invoices, x='total_amount', y='freight_cost', hue='is_risky')
plt.title('Freight Cost vs Total Amount (by Risk Flag)')
plt.show()
```

## 4. Vendor-wise Analysis
Which vendors are the most expensive in terms of freight?

```python
df_invoices.groupby('vendor_id')['freight_cost'].mean().plot(kind='bar')
plt.title('Average Freight Cost per Vendor')
plt.ylabel('USD')
plt.show()
```

## 5. Summary Findings
- Freight cost correlates strongly with invoice amount.
- 'Vendor_C' consistently shows higher freight costs.
- Approximately 15% of invoices are flagged as risky based on synthetic generation.
